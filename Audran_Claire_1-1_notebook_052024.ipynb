```python
def prediction_billet_logistic(data):
    data = data.set_index(data["id"])
    data.drop(["id", "diagonal", "height_left"], axis=1, inplace=True)
    model = joblib.load('logistic_reg_model.joblib')
    predictions = model.predict(data)
    data["is_genuine"] = predictions
    data["is_genuine"].replace([1, 0], [True, False], inplace=True)
    display(data)
```
